```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import skimage
from skimage import io
import glob
import cv2
import PIL
from PIL import Image
```

```python
datagen_br = ImageDataGenerator(
        brightness_range=[1.5,2.0] 
)

datagen_rot = ImageDataGenerator(
        rotation_range=45 
)

datagen_flip = ImageDataGenerator(
        horizontal_flip=True
)

datagen_ver = ImageDataGenerator(
        vertical_flip=True
)

datagen_dark = ImageDataGenerator(
        brightness_range=[0.05,0.2]
)
```

```python
closed_data= []
closed_file = glob.glob(r"C:/Users/User/Desktop/sem1_year4/fyp_thing/Testing/closed_mouth/*.PNG")
for my_closed_file in closed_file:
    print(my_closed_file)
    
    cld = Image.open(my_closed_file).convert('RGB')
    cld = np.array(cld)
    cld = cv2.resize(cld, (224,224))
    ##image = image/255.0
    closed_data.append(cld)

print('closed_mouth shape:', np.array(closed_data).shape)
```

```python
open_data= []
open_file = glob.glob(r"C:/Users/User/Desktop/sem1_year4/fyp_thing/Testing/open_mouth/*.PNG")
for my_open_file in open_file:
    print(my_open_file)
    
    opn = Image.open(my_open_file).convert('RGB')
    opn = np.array(opn)
    opn = cv2.resize(opn, (224,224))
    ##image = image/255.0
    open_data.append(opn)

print('open_mouth shape:', np.array(open_data).shape)
```

```python
for i in range(len(closed_data)):
    pic = closed_data[i]
    pic = pic.reshape((1,) + pic.shape)
    j = 0
    for batch in datagen_br.flow(pic,batch_size=16,
                         save_to_dir='aug_test/closed_mouth',
                         save_prefix='br',
                         save_format='PNG'):
        j += 1
        if j > 0:
            break
```
